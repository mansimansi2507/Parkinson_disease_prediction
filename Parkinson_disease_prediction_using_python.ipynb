```python
# drop unused columns
x=dataset.drop(columns=['status','name'],axis=1)
y=dataset['status']
```

```python
# Because data is not balanced so balance the data by undersampling
from imblearn.under_sampling import RandomUnderSampler  

under_sampler = RandomUnderSampler()
X_res,y_res= under_sampler.fit_resample(x,y)
```

```python
X_res.shape
```

```python
y_res.shape
```

```python
# cheak shape of original dataset and resampled dataset
from collections import Counter
print('Original dataset shape {}'.format(Counter(dataset['status'])))
print('Resampled dataset shape {}'.format(Counter(y_res)))
```

```python
#visualize resampled dataset
sns.countplot(y_res)
```

```python
# visualize by boxplot
i=1
plt.figure(figsize=(40,40))
for c in X_res.columns[:]:
    plt.subplot(6,4,i)
    plt.yticks(fontsize=15)
    plt.title(f"Boxplot of {c}",fontsize=15)
    sns.boxplot(data=X_res[c],color='blue')
    i+=1
plt.show()    
```

```python
#visualize by histogram
i=1
plt.figure(figsize=(40,40))
for c in X_res.columns[:]:
    plt.subplot(6,4,i)
    plt.yticks(fontsize=15)
    plt.title(f"Histogram of {c}",fontsize=15)
    plt.hist(X_res[c],color='blue')
    i+=1
plt.show()    
```

```python
#visualize by scatterplot
i=1
plt.figure(figsize=(40,40))
for c in X_res.columns[:]:
    plt.subplot(6,4,i)
    plt.yticks(fontsize=15)
    plt.title(f"Scatterplot of {c}",fontsize=15)
    plt.scatter(X_res[c],y_res,color='blue')
    i+=1
plt.show()    
```

```python
#split dataset into train and test data
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.20,random_state=42)
```

```python
#shape of resulting train and test data
print(X_res.shape, X_train.shape, X_test.shape)
```

```python
#standardization of data
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit(X_train)
```

```python
X_train=ss.transform(X_train)
X_test=ss.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
#Apply logistic regression technique 
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression(random_state=0)
```

```python
log_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
log_X_train_prediction=log_model.predict(X_train)
log_training_data_accuracy= accuracy_score(log_X_train_prediction, y_train)
```

```python
print("Accuracy on training data:",log_training_data_accuracy * 100)
```

```python
log_X_test_prediction=log_model.predict(X_test)
log_test_data_accuracy= accuracy_score(log_X_test_prediction, y_test)
```

```python
print("Accuracy on test data:", log_test_data_accuracy * 100)
```
